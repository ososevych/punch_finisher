```python
import pandas as pd
import numpy as np
import joblib
```

```python
model = joblib.load("model/rf_model.pkl")
scaler = joblib.load("model/scaler.pkl")
feature_names = joblib.load("model/feature_names.pkl")
```

```python
df = pd.read_csv("data/sample_input.csv")
df
```

```python
df['log_goal'] = np.log1p(df['goal'])
```

```python
df
```

```python
df.drop(['id', 'name', 'launched', 'deadline', 'pledged', 'state', 'goal', 'backers'], axis=1, inplace=True)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=["category", "subcategory", "country"])
```

```python
df
```

```python
for col in feature_names:
    if col not in df.columns:
        df[col] = 0
```

```python
df = df[feature_names]
```

```python
df_scaled = scaler.transform(df)
```

```python
pred = model.predict(df_scaled)[0]
proba = model.predict_proba(df_scaled)[0][1]
```

```python
print("Prediction:", "Successful" if pred == 1 else "Failed")
print("Probability of Success:", round(proba, 3))
```
